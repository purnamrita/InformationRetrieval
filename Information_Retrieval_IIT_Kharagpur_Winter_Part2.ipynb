```python
from sklearn.feature_extraction.text import TfidfVectorizer
c1 = 'In computer science artificial intelligence sometimes called machine intelligence is intelligence demonstrated by machines'
c2 = 'Experimentation calculation and Observation is called science'
c3 = 'Physics is a natural science that involves the study of matter and its motion through space and time, along with related concepts such as energy and force'
c4 = 'In mathematics and computer science an algorithm is a finite sequence of well-defined computer-implementable instructions'
c5 = 'Chemistry is the scientific discipline involved with elements and compounds composed of atoms, molecules and ions'
c6 = 'Biochemistry is the branch of science that explores the chemical processes within and related to living organisms'
c7 = 'Sociology is the study of society, patterns of social relationships, social interaction, and culture that surrounds everyday life'
corpus = [c1, c2, c3, c4, c5 ,c6, c7]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names())
print(X.shape)
```

```python
import pandas as pd
vector = X
df1 = pd.DataFrame(vector.toarray(), columns=vectorizer.get_feature_names())
df1
```

```python
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
print(stopwords.words('english'))
stop_words = set(stopwords.words('english'))
```

```python
def get_tokenized_list(doc_text):
    tokens = nltk.word_tokenize(doc_text)
    return tokens

def word_stemmer(token_list):
  ps = nltk.stem.PorterStemmer()
  stemmed = []
  for words in token_list:
    stemmed.append(ps.stem(words))
  return stemmed
```

```python
def remove_stopwords(doc_text):
  cleaned_text = []
  for words in doc_text:
    if words not in stop_words:
      cleaned_text.append(words)
  return cleaned_text
```

```python
cleaned_corpus = []
for doc in corpus:
  tokens = get_tokenized_list(doc)
  doc_text = remove_stopwords(tokens)
  doc_text  = word_stemmer(doc_text)
  doc_text = ' '.join(doc_text)
  cleaned_corpus.append(doc_text)
cleaned_corpus
```

```python
vectorizerX = TfidfVectorizer()
vectorizerX.fit(cleaned_corpus)
doc_vector = vectorizerX.transform(cleaned_corpus)
print(vectorizerX.get_feature_names())

print(doc_vector.shape)
```

```python
df1 = pd.DataFrame(doc_vector.toarray(), columns=vectorizerX.get_feature_names())
df1
```

```python
query = 'computer science'
query = get_tokenized_list(query)
query = remove_stopwords(query)
q = []
for w in word_stemmer(query):
  q.append(w)
q = ' '.join(q)
q
query_vector = vectorizerX.transform([q])
print(query_vector)
```

```python
# calculate cosine similarities
from sklearn.metrics.pairwise import cosine_similarity
cosineSimilarities = cosine_similarity(doc_vector,query_vector).flatten()
```

```python
related_docs_indices = cosineSimilarities.argsort()[:-10:-1]
print(related_docs_indices)

for i in related_docs_indices:
    data = [cleaned_corpus[i]]
    print(data)
```
